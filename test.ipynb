```python
import os
from dotenv import load_dotenv

load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
# prompt + llm + output


#prompt
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    # [ (시스템), (질문을 처리할 메세지) ]
    [ 
     ("system", "당신은 천문학 전문가입니다. 그리고 당신은 한국어로만 답변하는 AI입니다. "),
     ("user", "{input}")
     ]
)
```

```python
prompt_text = prompt.format(input="지구의 자전주기는 얼마인가요?")
```

```python
# from langchain_openai import ChatOpenAI

# # model 버전 명시하는게 좋음
# llm = ChatOpenAI(api_key=OPENAI_API_KEY, model_name="gpt-3.5-turbo-0125")


# OpenAI 유료로 인해 무료버전 진행
from langchain_ollama import ChatOllama

llm = ChatOllama(model="llama3:8b")
```

```python
response = llm.invoke(prompt_text)
```

```python
# content (답변) 속성만 추출
response.content
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
```

# LCEL
(입력, 모델, 출력 하나로 연결)

```python
# 입력, 모델, 출력 하나로 연결
chain = prompt | llm | output_parser
```

```python
response = chain.invoke({"input":"지구의 자전주기는 얼마인가요?"})
```
